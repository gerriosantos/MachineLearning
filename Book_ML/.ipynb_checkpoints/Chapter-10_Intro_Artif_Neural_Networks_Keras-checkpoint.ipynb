## Dos neurônios biológicos aos artificiais

- Surpreendentemente, as RNAs já existem há algum tempo: foram **introduzidas pela primeira vez em 1943** pelo neurofisiologista Warren McCulloch e pelo matemático Walter Pitts. Em seu artigo de referência, “Logical Calculus of Ideas Immanent in Nervous Activity”, McCulloch e Pitts apresentaram um modelo computacional simplificado de como os neurônios biológicos podem trabalhar juntos em cérebros de animais para realizar cálculos complexos usando lógica proposicional. Essa foi a primeira arquitetura de rede neural artificial. Desde então, muitas outras arquiteturas foram inventadas, como veremos.
  - Os sucessos iniciais das RNAs até a década de 1960 levaram à crença generalizada de que em breve estaríamos conversando com máquinas verdadeiramente inteligentes. Quando ficou claro que essa promessa não seria cumprida (pelo menos por um bom tempo), o financiamento foi para outro lugar e RNAs entraram em um longo inverno. 


- No **início dos anos 80**, houve um renascimento do interesse no **_conexionismo_** (o estudo de redes neurais), à medida que novas arquiteturas foram inventadas e melhores técnicas de treinamento foram desenvolvidas. Mas o progresso foi lento e, na década de 1990, outras poderosas técnicas de **_Machine Learning_** foram inventadas, como o **Support Vector Máquinas** (consulte o *Capítulo 5*). Essas técnicas pareciam oferecer melhores resultados e bases teóricas mais fortes do que as RNAs; portanto, mais uma vez o estudo das redes neurais entrou em um longo inverno.


- Finalmente, agora estamos testemunhando mais uma onda de interesse nas RNAs. Essa onda desaparecerá como as anteriores? Bem, existem algumas boas razões para acreditar que essa onda é diferente e que terá um impacto muito mais profundo em nossas vidas:

  - Atualmente, existe uma enorme quantidade de dados disponíveis para treinar redes neurais, e as RNAs freqüentemente superam outras técnicas de ML em problemas muito grandes e complexos.
  
  - O tremendo **aumento no poder de computação desde os anos 90** agora possibilita o treinamento de grandes redes neurais em um período de tempo razoável. Isso se deve em parte à **_Lei de Moore_**, mas também graças à **_indústria de jogos_**, que produziu milhões de placas de GPU poderosas.
  
  - Os algoritmos de treinamento foram aprimorados. Para ser justo, eles são apenas ligeiramente diferentes dos usados nos anos 90, mas esses ajustes relativamente pequenos têm um enorme impacto positivo.
  
  - Algumas limitações teóricas das RNAs se mostraram benignas na prática. Por exemplo, muitas pessoas pensaram que os algoritmos de treinamento da RNA estavam condenados porque provavelmente ficariam presos a **_ótimos locais_**, mas acontece que isso é bastante raro na prática (ou quando é o caso, eles geralmente são bastante próximos do **_ótimo global_**).
  
  - As **RNAs** parecem ter entrado em um círculo virtuoso de financiamento e progresso. Produtos surpreendentes baseados em RNAs fazem regularmente as manchetes, que atraem cada vez mais atenção e financiamento para eles, resultando em cada vez mais progresso e produtos ainda mais incríveis.
  

## Neurônios biológicos

- Antes de discutirmos os neurônios artificiais, vamos dar uma olhada rápida em um neurônio biológico (representado na **Figura 10-1**). 

  - É uma célula de aparência incomum, encontrada principalmente em **córtex cerebral de animais** (por exemplo, seu cérebro), composta por um **_corpo celular_** que contém o núcleo e a maioria dos componentes complexos da célula, além de muitas extensões ramificadas chamadas **_dendritos_**, além de uma extensão muito longa chamada **_axônio_**.   - Neurônios biológicos recebem impulsos elétricos curtos chamados **_sinais_** de outros neurônios por meio dessas sinapses. Quando um neurônio recebe um número suficiente de sinais de outros neurônios em alguns milissegundos, ele dispara seus próprios sinais.

## Computações Lógicas com Neurônios 

- **_Warren McCulloch_** e **_Walter Pitts_** propuseram um modelo muito simples do neurônio biológico, que mais tarde ficou conhecido como **neurônio artificial**: ele possui uma ou mais *inputs* binárias (on/off) e uma *output* binário. O neurônio artificial simplesmente ativa sua saída quando mais de um certo número de *input* está ativo. McCulloch e Pitts mostraram que, mesmo com um modelo tão simplificado, é possível construir uma rede de neurônios artificiais que calcula qualquer proposição lógica que você deseja. Por exemplo, vamos construir algumas **RNAs** que executam vários cálculos lógicos (veja a **Figura 10-3**), assumindo que um neurônio seja ativado quando pelo menos dois de seus *inputs* estiverem ativos.


```python
Image('fig10-3.png')
```

- A **primeira rede** à esquerda é simplesmente a função de identidade: se o neurônio **A** estiver ativado, o neurônio **C** também será ativado (uma vez que recebe dois sinais de *output* do neurônio **A**), mas se o neurônio **A** estiver desativado, o neurônio **C** estará desativado também.


- A **segunda rede** executa um termo lógico **AND**: o neurônio **C** é ativado apenas quando os neurônios **A** e **B** são ativados (um único sinal de entrada não é suficiente para ativar o neurônio **C**).


- A **terceira rede** executa um termo lógico **OR**: o neurônio **C** é ativado se o neurônio **A** ou o neurônio **B** estiver ativado (ou ambos).


- Finalmente, se supusermos que uma conexão de *input* possa inibir a atividade do neurônio (que é o caso dos neurônios biológicos), a **quarta rede** computará proposições lógicas ligeiramente mais complexas: o neurônio **C** é ativado apenas se o neurônio **A** estiver ativo e se o neurônio **B** estiver desativado. Se o neurônio **A** estiver ativo o tempo todo, você receberá um termo lógico **NOT**: o neurônio **C** estará ativo quando o neurônio **B** estiver desativado e vice-versa.


- Você pode imaginar facilmente como essas redes podem ser combinadas para calcular expressões lógicas complexas (**consulte os exercícios no final do capítulo**).

# O Perceptron

- O **Perceptron** é uma das arquiteturas mais simples da *RNA*, inventada em 1957 por **Frank Rosenblatt**. Baseia-se em um neurônio artificial levemente diferente (veja a **Figura 10-4**) chamado de **unidade lógica de limiar** (**Threshold Logic Unit - TLU** ) ou, às vezes, uma **unidade de limiar linear** (**Linear Threshold Unit - LTU**) : os *outputs* e os *inputs* agora são números (em vez de valores binários de *ativação/desativação*) e cada *input* com conexão está associada a um peso. A TLU calcula uma soma ponderada de seus *inputs* $z = w_{1} x_{1} + w_{2} x_{2} + ... + w_{n} x_{n} = \mathbf{x^{T}w}$, então aplica a **step function** a esta soma e gera o resultado: $h_{\mathbf{w}}(\mathbf{x}) = step(z)$, onde $z = \mathbf{x^{T}w}$.

```python
Image('fig10-4.png')
```

- A **step function** mais comum usada no Perceptrons é a **Heaviside step function** (consulte a **Equação 10-1**). Às vezes, a __função de sinal__ é usada.

```python
Image('eq10-1.png')
```

- Uma única **unidade lógica de limiar** (**TLU**) pode ser usada para *classificação binária linear simples*. Ele calcula uma combinação linear dos inputs e, se o resultado exceder um **limiar**, gera a classe positiva ou gera a classe negativa (assim como um *classificador de Regressão Logística ou um SVM linear*). 

  - Por exemplo, você pode usar uma única TLU para classificar flores de íris com base no comprimento e na largura da pétala (adicionando também um **feature** extra de **viés** $x_{0}=1$, como fizemos nos capítulos anteriores). Treinar uma TLU nesse caso significa encontrar os valores corretos para $w_{0}$, $w_{1}$ e $w_{2}$ (o algoritmo de treinamento será discutido em breve).
  
  
- Um **Perceptron** é simplesmente composto de uma única camada de TLUs, com cada TLU conectado a todas as **inputs**. Quando todos os neurônios de uma camada estão conectados a todos os neurônios da camada anterior (isto é, seus neurônios **inputs**), isso é chamado de **camada conectada completamente** (*fully connected layer*) ou **camada densa** (*dense layer*). 
  - Para representar o fato de que cada **input** é enviada para cada TLU, é comum desenhar neurônios de passagens especiais chamados **neurônios de input** (*input neurons*): eles apenas produzem qualquer *input* que são alimentados. Todos os *neurônios de inputs* formam a *camada de input*. Além disso, um *feature* de viés extra é geralmente adicionado ($x_{0} = 1$): ele é normalmente representado usando um tipo especial de neurônio chamado **viés de neurônio** (**bias neuron**), que gera apenas $1$ o tempo todo. Um Perceptron com dois **inputs** e três **outputs** é representado na **Figura 10-5**. Esse Perceptron pode classificar instâncias simultaneamente em três classes binárias diferentes, o que o torna um classificador de vários **outputs**.

```python
Image('fig10-5.png')
```

- Graças à magia da álgebra linear, é possível calcular com eficiência os **outputs** de uma camada de neurônios artificiais para *várias instâncias* de uma só vez, usando a **Equação 10-2**:

     - Equation 10-2. Computing the outputs of a fully connected layer -
        
        $$h_{\mathbf{W,b}}(\mathbf{X}) = \phi(\mathbf{XW + b})$$

- Como sempre, $\mathbf{X}$ representa a matriz de input **features**. Possui uma linha por **instância**, uma coluna por **feature**.


- A matriz de pesos $\mathbf{W}$ contém todos os pesos de conexão, exceto neurônio viés. Possui uma linha por neurônio de input e uma coluna por neurônio artificial na camada.


- O vetor de viés $\mathbf{b}$ contém todos os pesos de conexão entre o neurônio de viés e os neurônios artificiais. Ele tem um termo de viés por neurônio artificial.


- A função ${\phi}$ é chamada de **função de ativação**: quando os neurônios artificiais são TLUs, é uma **step function** (mas discutiremos outras funções de ativação em breve).


- Então, como um Perceptron é treinado? 

  - O algoritmo de treinamento Perceptron proposto por **Frank Rosenblatt** foi amplamente inspirado na **regra de Hebb**. Em seu livro **The Organization of Behavior**, publicado em 1949, **Donald Hebb** sugeriu que, quando um neurônio biológico geralmente aciona outro neurônio, a conexão entre esses dois neurônios fica mais forte. 
  - Mais tarde, essa ideia foi resumida por **Siegrid Löwel** nesta frase cativante: "Células que disparam juntas, conectam-se". Essa regra mais tarde ficou conhecida como **regra de Hebb** (ou **aprendizado de Hebbian**); isto é, o peso da conexão entre dois neurônios aumenta sempre que eles têm a mesma saída. 
  - Perceptrons são treinados usando uma variante desta regra que leva em consideração o erro cometido pela rede; ele reforça conexões que ajudam a reduzir o erro. Mais especificamente, o Perceptron é alimentado com uma instância de treinamento por vez e, para cada instância, faz suas previsões. Para cada produto do neurônio que produziu uma previsão errada, ele reforça os pesos de conexão dos inputs que teriam contribuído para a previsão correta. A regra é mostrada na **Equação 10-3**.
  
  
  Equation 10-3. Perceptron learning rule (weight update)
  
  
  $$ w_{i,j}^{(next\_step)} = w_{i,j} + \eta(y_{j} - \hat{y_{j}})x_{i} $$
  
  
  - $w_{i,j}$ é a conexão de peso entre o i-ésimo neurônio de input e o j-ésimo nerônio de produto.
  
  
  - $x_{i}$ é o i-ésimo valor do input da instância de treinamento atual.
  
  
  - $\hat{y_{j}}$ é o output do j-ésimo neurênio de output para instância de treinamento atual.
  
  
  - $y_{j}$ é o target output do j-ésimo neurônio de output para instância de treinamento atual.
  
  
  - $\eta$ é a taxa de aprendizagem.

  

- O limite de decisão de cada neurônio de output é linear, portanto os Perceptrons são incapazes de aprender padrões complexos (assim como os classificadores de Regressão Logística). No entanto, se as instâncias de treinamento forem linearmente separáveis, Rosenblatt demonstrou que esse algoritmo convergiria para uma solução. Isso é chamado de **teorema da convergência de Perceptron**.


- O **Scikit-Learn** fornece uma classe Perceptron que implementa uma única rede TLU. Ele pode ser usado da maneira que você esperaria - por exemplo, no conjunto de dados da íris (apresentado no capítulo 4):
  

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
X = iris.data[:, (2, 3)] # petal length, petal width
y = (iris.target == 0).astype(np.int) # Iris Setosa?
per_clf = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
per_clf.fit(X, y)

y_pred = per_clf.predict([[2, 0.5]])
```

```python
y_pred = per_clf.predict([[2, 0.5]])
```

```python
y_pred
```

```python
a = -per_clf.coef_[0][0] / per_clf.coef_[0][1]
b = -per_clf.intercept_ / per_clf.coef_[0][1]

axes = [0, 5, 0, 2]

x0, x1 = np.meshgrid(
        np.linspace(axes[0], axes[1], 500).reshape(-1, 1),
        np.linspace(axes[2], axes[3], 200).reshape(-1, 1),
    )
X_new = np.c_[x0.ravel(), x1.ravel()]
y_predict = per_clf.predict(X_new)
zz = y_predict.reshape(x0.shape)

plt.figure(figsize=(10, 4))
plt.plot(X[y==0, 0], X[y==0, 1], "bs", label="Not Iris-Setosa")
plt.plot(X[y==1, 0], X[y==1, 1], "yo", label="Iris-Setosa")

plt.plot([axes[0], axes[1]], [a * axes[0] + b, a * axes[1] + b], "k-", linewidth=3)
from matplotlib.colors import ListedColormap
custom_cmap = ListedColormap(['#9898ff', '#fafab0'])

plt.contourf(x0, x1, zz, cmap=custom_cmap)
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="lower right", fontsize=14)
plt.axis(axes)

#save_fig("perceptron_iris_plot")
plt.show()
```

- Você deve ter notado o fato de que o algoritmo de aprendizado Perceptron se assemelha fortemente à **Stochastic Gradient Descent**. De fato, a classe Perceptron do Scikit-Learn é equivalente a usar um **SGDClassifier** com os seguintes hiperparâmetros: loss = "perceptron", learning_rate = "constant", eta0 = 1 (taxa de aprendizado) e penalty=None (sem regularização).

  - Observe que, ao contrário dos classificadores de regressão logística, os Perceptrons não produzem uma probabilidade de classe; em vez disso, eles apenas fazem previsões com base em um limite rígido. Essa é uma das boas razões para preferir a regressão logística do que os Perceptrons.
  
  - Na **monografia de 1969 intitulada Perceptrons**, Marvin Minsky e Seymour Papert destacaram uma série de fraquezas sérias dos Perceptrons, em particular o fato de serem incapazes de resolver alguns problemas triviais (por exemplo, o problema de classificação OR Exclusivo (**XOR**); veja o lado esquerdo da Figura 10-6). Claro que isso é verdade para qualquer outro modelo de classificação linear também (como os classificadores de Regressão Logística), mas os pesquisadores esperavam muito mais dos Perceptrons, e sua decepção foi grande, e muitos pesquisadores abandonaram as redes neurais em favor de problemas de nível superior, como lógica, resolução de problemas e pesquisa.
  
  
  - No entanto, verifica-se que **algumas das limitações do Perceptrons podem ser eliminadas empilhando vários Perceptrons**. A RNA resultante é chamada de Perceptron de várias camadas (**Multi-Layer Perceptron - MLP**). Em particular, um MLP pode resolver o problema **XOR**, como você pode verificar computando o output do MLP representada à direita da **Figura 10-6**: com inputs (0,0) ou (1,1), a rede produz 0, e com as inputs (0,1) ou (1,0), ele gera 1. Todas as conexões têm um peso igual a 1, exceto as quatro conexões em que o peso é mostrado. Tente verificar se esta rede realmente resolve o problema do XOR!

# Perceptron Multicamadas e _Backpropagation_


- Um **MLP** é composto de uma camada de entrada (**passthrough**), uma ou mais camadas de **TLUs**, as chamadas **camadas ocultas**, e uma camada final de TLUs, chamada camada de saída (veja a **Figura 10-7**). As camadas próximas à camada de entrada são geralmente chamadas de camadas inferiores e as próximas às saídas são chamadas de camadas superiores. Cada camada, **EXCETO a camada de saída**, inclui um **neurônio de viés** e está totalmente conectada à próxima camada.

```python
Image('fig10-7.png')
```

   - O sinal flui apenas em uma direção (das entradas às saídas); portanto, essa arquitetura é um exemplo de uma **rede neural _feedforward_** (**FNN**).


- Quando uma RNA contém uma bateria profunda de camadas ocultas, é chamada de **_deep neural network_** (**DNN**). O campo de Deep Learning estuda DNNs e, mais geralmente, modelos que contêm baterias profundas de cálculos. No entanto, muitas pessoas falam sobre Deep Learning sempre que redes neurais estão envolvidas (mesmo as rasas).


- Por muitos anos, os pesquisadores lutaram para encontrar uma maneira de treinar MLPs, sem sucesso. Mas, em **1986**, **David Rumelhart, Geoffrey Hinton e Ronald Williams** publicaram um artigo inovador introduzindo o algoritmo de treinamento de **backpropagation**, que ainda é usado hoje. Em suma, é simplesmente um  **gradiente descendente** (introduzida no **capítulo 4**) usando uma técnica eficiente para calcular os gradientes automaticamente: em apenas duas passagens pela rede (uma para frente, uma para trás), o algoritmo de **backpropagation** é capaz de calcular o gradiente do erro da rede em relação a cada modelo de parâmetro simples. Em outras palavras, ele pode descobrir como cada peso de conexão e cada termo de viés devem ser ajustados para reduzir o erro. Depois de ter esses gradientes, ele apenas executa uma etapa regular de gradiente descendente e todo o processo é repetido até a rede convergir para a solução.

   - Computar gradientes automaticamente é chamado de **diferenciação automática** ou **autodiff**. Existem várias técnicas de **autodiff**, com diferentes prós e contras. O usado pelo **backpropagation** é chamado de **_autodiff no modo reverso_**. É rápido e preciso, e é adequado quando a função diferenciada possui muitas variáveis (por exemplo, pesos de conexão) e poucas saídas (por exemplo, uma perda). Se você quiser saber mais sobre o autodiff, confira ???
   
   

- Vamos analisar esse algoritmo com mais detalhes:

   - Ele lida com um minilote (**_mini-batch_**) de cada vez (por exemplo, contendo 32 instâncias cada) e realiza o treinamento completo várias vezes. Cada passe é chamado de **época**, como vimos no Capítulo 4.
   
   - Cada mini-batch é passado para a camada de entrada da rede, que apenas a envia para a primeira camada oculta. O algoritmo calcula a saída de todos os neurônios nesta camada (para todas as instâncias do mini-lote). O resultado é passado para a próxima camada, sua saída é calculada e passada para a próxima camada, e assim sucessivamente até obtermos a saída da última camada, a camada de saída. Este é o passe para frente (**forward pass**): é exatamente como fazer previsões, exceto que todos os resultados intermediários são preservados, pois são necessários para a passagem para trás (**backward pass**).
   
   - Em seguida, o algoritmo mede o **erro de saída da rede** (ou seja, usa uma **função de perda** que compara a **saída prevista** e a **saída real** da rede e retorna alguma **medida do erro**).

   - Em seguida, calcula quanto cada conexão de saída contribuiu para o erro. Isso é feito analiticamente, basta aplicar a **regra da cadeia** (talvez a regra mais fundamental no cálculo), o que torna essa etapa rápida e precisa.

   - O algoritmo mede quanto dessas contribuições de erro veio de cada conexão na camada abaixo (**voltando**), novamente usando a regra da cadeia - e assim por diante até que o algoritmo atinja a camada de entrada. Como explicamos anteriormente, essa passagem reversa mede com eficiência o erro do gradiente em todos os pesos de conexão na rede, propagando o gradiente de erro para trás pela rede (daí o nome do algoritmo).
   
   - Por fim, o algoritmo executa uma etapa de **gradiente descendente** para ajustar todos os pesos de conexão na rede, usando os gradientes de erro que acabou de computar.



 



- Esse algoritmo é tão importante que vale a pena **resumir novamente**: para cada instância de treinamento, o algoritmo de **backpropagation** primeiro faz uma previsão (passagem para frente - **forward pass**), mede o erro e depois passa por cada camada no sentido inverso para medir a contribuição do erro de cada conexão (passagem reversa - **reverse pass**) e, finalmente, ajusta ligeiramente a conexão pesos para reduzir o erro (etapa do gradiente descendente).

   - É importante inicializar **todos os pesos de conexão das camadas ocultas aleatoriamente**, caso contrário, o treinamento falhará. Por exemplo, se você inicializar todos os pesos e vieses para zero, todos os neurônios em uma determinada camada serão perfeitamente idêntico e, dessa forma, a **retropropagação** os afetará exatamente da mesma maneira, portanto eles permanecerão idênticos. Em outras palavras, apesar de ter centenas de neurônios por camada, seu modelo agirá como se tivesse apenas um neurônio por camada: não será muito inteligente. Se, em vez disso, você inicializar aleatoriamente os pesos, você quebrará a simetria e permitirá a **retropropagação** para treinar uma equipe diversificada de neurônios.

- Para que esse algoritmo funcionasse corretamente, os autores fizeram uma alteração importante na arquitetura do **perceptron multicamdas**: **_substituíram a função step_** pela **função logística**, $\sigma(z) = \frac{1}{1+e^{-z}}$. 

   - Isso foi essencial porque a _função step_ **_contém apenas segmentos planos_**, portanto, **não há gradiente com o qual trabalhar** (o gradiente descendente não pode se mover em uma superfície plana), enquanto a _função logística_ **_possui uma derivada diferente de zero bem definida em todos os lugares_**, permitindo que o gradiente descendente faça algum progresso a cada passo. De fato, o algoritmo de **retropropagação** funciona bem com muitas outras funções de ativação, não apenas com a função logística. **Duas outras funções de ativação populares são**:
   
   - **A função tangente hiperbólica**  $tanh(z) = 2\sigma(2z) -1$ 
   
      - Assim como a função logística, é em forma de $S$, contínua e diferenciável, mas sua o valor da saída varia de $–1$ a $1$ (em vez de $0$ a $1$ no caso da *função logística*), o que tende a tornar a saída de cada camada mais ou menos centralizada em torno de $0$ no início do treinamento. Isso geralmente ajuda a acelerar a convergência.
      
   - A **função de Unidade linear retificada** (**The Rectified Linear Unit function**: $ReLU(z) = max(0, z)$    
      - É contínua, mas infelizmente não diferenciável em $z = 0$ (a inclinação muda abruptamente, o que pode fazer com que a descida do gradiente salte ao redor) e sua derivada é 0 para z <0. No entanto, na prática, funciona muito bem e tem a vantagem de ser rápido na computação 11. Mais importante ainda, o fato de ele não ter um valor máximo de saída também ajuda a reduzir alguns problemas durante a descida do gradiente (voltaremos a isso no **capítulo 11**).
      
      

- Essas funções **populares de ativação e suas derivadas** estão representados na **Figura 10-8**. Mas espere! Por que precisamos de funções de ativação em primeiro lugar? Bem, se você encadear várias transformações lineares, tudo o que você obtém é uma **transformação linear**. Por exemplo, diga $f(x) = 2 x + 3$ e $g(x) = 5 x - 1$, então encadear essas duas funções lineares fornece outra função linear: $f(g (x)) = 2 (5 x - 1 ) + 3 = 10 x + 1$. Portanto, se você não possui uma não linearidade entre as camadas, mesmo uma monte de camadas profunda é equivalente a uma única camada: você não pode resolver problemas muito complexos com isso.

# Regressão MPLs


- Primeiro, os MLPs podem ser usados para **tarefas de regressão**. Se você deseja prever um valor único (por exemplo, o preço de uma casa, considerando muitas de suas características), precisará apenas de um único neurônio de saída: sua saída é o valor previsto. Para a **regressão multivariada** (ou seja, para prever vários valores de uma vez), você precisa de um neurônio de saída por dimensão de saída (um vetor de saídas). Por exemplo, para localizar o centro de um objeto em uma imagem, é necessário prever coordenadas 2D e, portanto, dois neurônios de saída. Se você também deseja colocar uma caixa delimitadora em torno do objeto, precisará de mais dois números: a largura e a altura do objeto. Então você acaba com 4 neurônios de saída.


- Em geral, ao criar um MLP para regressão, você não deseja usar nenhuma função de ativação para os neurônios de saída; portanto, eles são livres para emitir qualquer intervalo de valores. No entanto, se você quiser garantir que a **saída sempre seja positiva**, poderá usar a **função de ativação ReLU** ou a **função de ativação softplus** na **camada de saída**. Por fim, se você quiser garantir que as previsões caiam dentro de um determinado intervalo de valores, poderá usar a **função logística** ou a **tangente hiperbólica** e dimensionar os rótulos para o intervalo apropriado: $0$ a $1$ para a função logística, ou $-1$ a $1$ para a tangente hiperbólica.


- A **função de perda** a ser usada durante o treinamento é tipicamente o **erro quadrático médio**, mas se você tiver muitos valores discrepantes no conjunto de treinamento, poderá preferir usar o **erro absoluto médio**. Como alternativa, você pode usar a **perda Huber**, que é uma combinação de ambas.

    - A **perda de Huber** é quadrática quando o erro é menor que o limiar $\delta$ (normalmente $1$), mas linear quando o erro for maior que $\delta$. Isso o torna menos sensível a valores discrepantes do que o erro médio quadrático e, geralmente, é mais preciso e converge mais rápido que o erro absoluto médio.

# Classificação de MPLs

- MLPs também podem ser usados para **tarefas de classificação**. 

    - Para um problema de **classificação binária**:
       - você só precisa de um único neurônio de saída usando a **função de ativação logística**: a saída será um número entre $0$ e $1$, que você pode **interpretar** como a **probabilidade estimada da classe positiva**. Obviamente, a probabilidade estimada da classe negativa é igual a um menos esse número.

    - As MLPs também podem lidar facilmente com tarefas de **classificação binária de várias rótulos**,(consulte o Capítulo 3): 
      - Por exemplo, você poderia ter um sistema de classificação de e-mail que prediz se cada e-mail recebido é **ham** ou **spam** e, simultaneamente, se é um e-mail **urgente** ou **não urgente**. Nesse caso, você precisaria de dois neurônios de saída, ambos usando a **função de ativação logística**:       
      
      - o primeiro geraria a probabilidade de o email ser spam e o segundo geraria a probabilidade de ser urgente. Em geral, você dedica um neurônio de saída para cada classe positiva. Observe que as probabilidades de saída não somam necessariamente um. Isso permite que o modelo produza qualquer combinação dos rótulos: você pode ter **ham não urgente**, **ham urgente**, **spam não urgente** e talvez até **spam urgente** (embora isso provavelmente seja um erro).
      
  - Para um problema de **classificação multiclasse**:
      - Se cada instância puder pertencer apenas a uma única classe, dentre 3 ou mais classes possíveis (por exemplo, classes 0 a 9 para classificação de imagem com dígitos), será necessário ter um neurônio de saída por classe e usar a função de ativação do softmax para toda a camada de saída (veja a Figura 10-9). A função softmax (apresentada no capítulo 4) garantirá que todas as probabilidades estimadas estejam entre $0$ e $1$ e que as soma dos valores das $9$ classes somam $1$ (que é necessária se as classes forem exclusivas). Isso é chamado de **classificação multiclasse**.
