- Uma tarefa relacionada é a **redução da dimensionalidade**, na qual o objetivo é simplificar os dados sem perder muitas informações. ### Aprendizagem Semissupervisionada

- Alguns algoritmos podem lidar com dados de treinamento **_parcialmente rotulados_**, geralmente **_muitos dados não rotulados e um pouco de dados rotulados_**. Isso é chamado de aprendizado semissupervisionado (Figura 1-11).
  - Uma combinação de **algoritmos de aprendizagens supervisionadas e Não-Supervisionadas**.
    - **Deep Belief Networks (DBNS)** -- Baseado em componentes Não-Supervisionados chamados **_Restricted Boltzmann Machines (RBMs)_**, empilhados em cima um do outro. 
      - **_RBM_** usa treinamento sequencial de Aprendizagem Não-Supervisionada, depois todo o sistema é ajustado com técnicas de Aprendizagem Supervisionada.
      
      
- Um exemplo é o serviço de hospedagem de fotos, *Google Fotos*. Depois de carregar todas as fotos de sua família no serviço, ele reconhece automaticamente que a mesma pessoa A aparece nas fotos 1, 5 e 11, enquanto outra pessoa B aparece nas fotos 2, 5 e 7. Essa é a parte **Não-Supervisionada do algoritmo (clustering)**. Agora, tudo o que o sistema precisa é que você diga quem são essas pessoas. Apenas um **_rótulo_** por pessoa, e é capaz de nomear todos em todas as fotos, o que é útil para pesquisar fotos.

### Aprenidzagem por Reforço

### Aprendizagem de Lote (Batch Learning)

- Deve ser treinado usando todos os dados disponíveis.
  - Normalmente, é feito *offline*  --- devido o custo computacional de tempo.
  - Primieiro é treinado, depois é executado aplicando apenas o que aprendeu (se aprender mais). 
  - Ele é chamado também de aprendizagem *offline*.

### Aprendizagem Online

- No **Aprendizado Online**, você treina o sistema de forma incremental, alimentando **_instâncias_** de dados sequencialmente, individualmente ou por pequenos grupos chamados **mini-lotes**. Cada etapa do aprendizado é rápida e barata, para que o sistema possa aprender sobre novos dados rapidamente, à medida que eles chegam (veja a Figura 1-13).

```python
Image("figura1-13.png")
```

- O **Aprendizado Online** é ótimo para sistemas que recebem dados como um fluxo contínuo (por exemplo, preços das ações) e precisam se adaptar às mudanças de forma rápida ou autônoma. Também é uma boa opção se você tiver recursos de computação limitados: depois que um sistema de aprendizado on-line aprende sobre novas instâncias de dados, ele não precisa mais delas, para que você possa descartá-las (a menos que queira reverter para uma versão anterior e "reproduzir" os dados). Isso pode economizar uma quantidade enorme de espaço.


- Os algoritmos de **Aprendizado Online** também podem ser usados para treinar sistemas em grandes conjuntos de dados que não cabem na memória principal de uma máquina (isso é chamado de aprendizado **_fora do núcleo_**). O algoritmo carrega parte dos dados, executa uma etapa de treinamento nesses dados e repete o processo até que seja executado em todos os dados (veja a Figura 1-14).
  
  - O aprendizado fora do núcleo geralmente é feito offline (ou seja, não no sistema ativo), portanto, o Aprendizado Online pode ser um nome confuso. Pense nisso como um **Aprendizado Incremental**.

```python
Image("figura1-14.png")
```

- Um parâmetro importante dos sistemas de **Aprendizado Online** é a rapidez com que eles devem se adaptar à alteração dos dados: isso é chamado de **_taxa de aprendizado_**. Se você definir uma alta taxa de aprendizado, seu sistema se adaptará rapidamente a novos dados, mas também tenderá a esquecer rapidamente os dados antigos (você não deseja que um filtro de spam sinalize apenas os tipos mais recentes de spam exibidos) . Por outro lado, se você definir uma baixa taxa de aprendizado, o sistema terá mais inércia; isto é, ele aprenderá mais devagar, mas também será menos sensível ao ruído nos novos dados ou às seqüências de pontos de dados não representativos (**_outliers_**).


- Um grande desafio do aprendizado on-line é que, se dados incorretos forem fornecidos ao sistema, o desempenho do sistema diminuirá gradualmente. Se estamos falando de um sistema ativo, seus clientes perceberão. Por exemplo, dados ruins podem vir de um sensor com defeito em um robô ou de alguém enviando um spam para um mecanismo de pesquisa para tentar obter uma classificação alta nos resultados de pesquisa. Para reduzir esse risco, você precisa monitorar seu sistema de perto e prontamente desativar o aprendizado (e possivelmente reverter para um estado anterior) se detectar uma queda no desempenho. Você também pode monitorar os dados de entrada e reagir a dados anormais (por exemplo, usando um algoritmo de detecção de anomalias).

## Aprendizagem Baseada em Instância vs Aprendizagem Baseada em Modelo

- Mais uma maneira de categorizar os sistemas de Machine Learning é como eles **generalizam**. A maioria das tarefas de Machine Learning é sobre fazer previsões. Isso significa que, dados vários exemplos de treinamento, o sistema precisa ser capaz de generalizar para exemplos que nunca viu antes. Ter uma boa medida de desempenho nos dados de treinamento é bom, mas insuficiente; o **verdadeiro objetivo** é ter um **bom desempenho em novas instâncias**.

- Existem duas abordagens principais para generalização: aprendizado baseado em instância e aprendizado baseado em modelo:

### Aprendizagem Baseada em Instância

- Possivelmente, a forma mais trivial de aprender é simplesmente memorizar. Se você criar um filtro de spam dessa maneira, apenas sinalizará todos os emails idênticos aos emails que já foram sinalizados pelos usuários --- não a pior solução, mas certamente não a melhor.


- Em vez de apenas sinalizar emails idênticos aos emails de spam já conhecidos, seu filtro de spam pode ser programado para sinalizar emails muito semelhantes aos emails de spam já conhecidos. Isso requer uma medida de semelhança entre dois emails. Uma medida de similaridade (muito básica) entre dois emails pode ser contar o número de palavras que eles têm em comum. O sistema sinalizaria um email como spam se tiver muitas palavras em comum com um email de spam já conhecido.


- Isso se chama aprendizado baseado em instâncias: o sistema aprende os exemplos memorizando, generalizando-os para novos casos, comparando-os com os exemplos aprendidos (ou um subconjunto deles), usando uma medida de similaridade. Por exemplo, (na Figura 1-15), a **_nova instância_** seria *classificada* como um triângulo porque a maioria das instâncias mais semelhantes pertence a essa **_classe_**.

```python
Image("figura1-15.png")
```

### Aprendizagem Baseada em Modelo

- Outra maneira de **generalizar** a partir de um conjunto de exemplos é construir um modelo desses exemplos e usá-lo para fazer previsões. Isso é chamado de **Aprendizado Baseado em Modelo** (Figura 1-16):
 

```python
Image("figura1-16.png")
```

- Por exemplo, suponha que você queira saber se o dinheiro deixa as pessoas felizes e faça o *download* dos dados do *Índice de Bem-Estar* no *site* da *OCDE*, bem como estatísticas sobre o PIB *per capita* no *site* do *FMI*. Então você junta às tabelas e classifica por PIB *per capita*. A Tabela 1-1 mostra um trecho do que você recebe.

```python
Image("table1.png")
```

- Vamos traçar os dados para alguns países aleatórios (Figura 1-17).

```python
Image("figura1-17.png")
```

- Parece haver uma tendência aqui! Embora os dados sejam ruídos (ou seja, parcialmente aleatórios), parece que a satisfação com a vida aumenta mais ou menos linearmente à medida que o PIB *per capita* do país aumenta. Então você decide modelar a satisfação com a vida como uma função linear do PIB *percapita*. Essa etapa é chamada seleção de modelo: você selecionou um **modelo linear** de satisfação da vida com apenas um **_atributo_**, o PIB *per capita* (Equação 1-1).


   - Equação: **Modelo Linear Simples**
   
      $life\_satisfaction = \theta_{0} + \theta_{1}GDP\_per\_capita$

- Este modelo possui dois parâmetros de modelo, $\theta_{0}$ e $\theta_{1}$ . Ajustando esses parâmetros, você pode fazer com que seu modelo represente qualquer função linear, como mostra a Figura 1-18.

```python
Image("figura1-18.png")
```

- Antes de usar o modelo -- definir $\theta_{0}$ e $\theta_{1}$. Ou seja, com que valores de $\theta_{0}$ e $\theta_{1}$ o modelo tem melhor desempenho?
  - Precisa saber uma medidade de desempenho para o modelo.
    - Pode usar a **Função de Utilidade** --- para saber o quão **bom** é o modelo.
    - Pode usar a **Função Custo** --- para saber o quão **ruim** é o modelo


- Para esse modelo, geralmente, usa-se a **Função Custo**, buscando minimizar a distância entre as previsões do modelo linear os exemlplos de treinamento.


- É aqui que entra o **Algoritmo de Regressão Linear**: você fornece seus exemplos de **_treinamento_** e encontra os parâmetros que tornam o modelo linear mais adequado aos seus dados. Isso é chamado de **_treinamento do modelo_**. 


- Agora, o modelo ajusta os dados de treinamento o mais próximo possível (para um modelo linear), conforme você pode ver na (Figura 1-19).

```python
Image("figura1-19.png")
```

- Você está finalmente pronto para executar o modelo para fazer previsões. Por exemplo, digamos que você queira saber como os cipriotas estão felizes e que os dados da OCDE não têm a resposta. Felizmente, você pode usar seu modelo para fazer uma boa previsão: você consulta o PIB *per capita* de Chipre, encontra $22.587$ dolares, aplica seu modelo e descobre que a satisfação com a vida provavelmente está em torno de $4,85 + 22,587 × 4,91 × 10^{-5} = 5.96$


- O exemplo 1-1 mostra o código Python que carrega os dados, os prepara, cria um gráfico de dispersão para visualização e, em seguida, treina um modelo linear e faz uma previsão.


**Exemplo 1-1.** Treinamento e execução de um modelo linear usando o *Scikit-Learn*

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model
```

```python
# Baixar dados -- Dad
## Local do github, onde está o banco de dados.
       
felic = "https://raw.githubusercontent.com/ageron/handson-ml2/master/datasets/lifesat/oecd_bli_2015.csv"
oecd_bli = pd.read_csv(felic, thousands=',', delimiter=',')
oecd_bli.head()
```

```python
len(oecd_bli)
```

```python
## Baixar dados

GDP = "https://raw.githubusercontent.com/ageron/handson-ml2/master/datasets/lifesat/gdp_per_capita.csv"
#GDP = "https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/lifesat/gdp_per_capita.csv"
gdp_per_capita = pd.read_csv(GDP, thousands=',', delimiter='\t',
                            encoding='latin1', na_values='n/a')
gdp_per_capita.head()
```

```python
#gdp_per_capita.rank()
len(gdp_per_capita)
```

```python
## Função encontrada em forum -- https://github.com/ageron/handson-ml/issues/33
    ## Essa função não tem no book.
    
def prepare_country_stats(oecd_bli, gdp_per_capita):
    oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
    oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
    gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
    gdp_per_capita.set_index("Country", inplace=True)
    full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,
                                  left_index=True, right_index=True)
    full_country_stats.sort_values(by="GDP per capita", inplace=True)
    remove_indices = [0, 1, 6, 8, 33, 34, 35]
    keep_indices = list(set(range(36)) - set(remove_indices))
    return full_country_stats[["GDP per capita", 'Life satisfaction']].iloc[keep_indices]
```

```python
## Preparar os dados
country_stats = prepare_country_stats(oecd_bli, gdp_per_capita)
X = np.c_[country_stats["GDP per capita"]]
y = np.c_[country_stats["Life satisfaction"]]
```

```python
## Visualizar os dados

country_stats.plot(kind='scatter', x="GDP per capita", y="Life satisfaction")
plt.show()
```

```python
# Seleionando um modelo linear
model = sklearn.linear_model.LinearRegression()
```

```python
# treinando o modelo

model.fit(X,y)
```

```python
## Fazer a predição para Chipre
X_new = [[22587]]
```

```python
print(model.predict(X_new))
```

- Se você tivesse usado um algoritmo de aprendizado baseado em instâncias, descobriria que a Eslovênia tem o PIB per capita mais próximo do de Chipre (USS 20.732) e, como os dados da *OCDE* nos dizem que a satisfação de vida dos eslovenos é 5,7, você teria previsto uma satisfação de vida de 5,7 para Chipre. Se você diminuir um pouco o zoom e observar os dois próximos países mais próximos, encontrará Portugal e Espanha com satisfações de vida de 5,1 e 6,5, respectivamente. Na média desses três valores, você obtém 5,77, o que é bastante próximo da previsão baseada em modelo. Esse algoritmo simples é chamado de **regressão k-vizinhos mais próximos** (neste exemplo, k = 3). 

- Substituindo o **modelo de regressão linear** pela **regressão dos k-vizinhos mais próximos**, no código anterior é tão simples quanto substituir esses dois linhas:

```python
import sklearn.neighbors

model_1 = sklearn.neighbors.KNeighborsRegressor(n_neighbors=3)
```

```python
model_1.fit(X, y)
```

### Principais desafios do aprendizado de máquina

- Em resumo, como sua tarefa principal é selecionar um algoritmo de aprendizado e treiná-lo em alguns dados, as duas coisas que podem dar errado são **"algoritmo ruim"** e **"dados ruins"**. Vamos começar com exemplos de dados ruins.


### Quantidade insuficiente de dados de treinamento


- Para uma criança aprender o que é uma maçã, basta apontar para uma maçã e dizer “maçã” (possivelmente repetindo esse procedimento algumas vezes). Agora a criança é capaz de reconhecer maçãs em todos os tipos de cores e formas. Gênio.


- O aprendizado de máquina ainda não chegou a isto; são necessários muitos dados para que a maioria dos algoritmos de **Machine Learning** funcione corretamente. Mesmo para problemas muito simples, você normalmente precisa de milhares de exemplos, e para problemas complexos, como reconhecimento de imagem ou fala, você pode precisa de milhões de exemplos (a menos que possa reutilizar partes de um modelo existente).


### Dados de treinamento não representativos

- Para generalizar bem, é crucial que seus dados de treinamento sejam representativos dos novos casos para os quais você deseja generalizar. Isso é verdade se você usa aprendizado **baseado em instância** ou **aprendizado baseado em modelo**.
  - Por exemplo, o conjunto de países que usamos anteriormente para treinar o modelo linear não era perfeitamente representativo; alguns países estavam desaparecidos. A (Figura 1-21) mostra como são os dados quando você adiciona os países ausentes. 

```python
Image("figura1-21.png")
```

- Se você treina um modelo linear nesses dados, obtém a linha sólida, enquanto o modelo antigo é representado pela linha pontilhada. Como você pode ver, a adição de alguns países ausentes não apenas altera significativamente o modelo, mas também deixa claro que um modelo linear tão simples provavelmente nunca funcionará bem. Parece que os países muito ricos não são mais felizes do que os países moderadamente ricos (de fato, parecem mais infelizes) e, inversamente, alguns países pobres parecem mais felizes do que muitos países ricos. 


- Usando um conjunto de treinamento não representativo, treinamos um modelo que dificilmente fará previsões precisas, especialmente para países muito pobres e muito ricos.


-  É crucial usar um conjunto de treinamento representativo dos casos para os quais você deseja generalizar. Isso geralmente é mais difícil do que parece: se a amostra for muito pequena, você terá ruído de amostragem (ou seja, dados não representativos como resultado do acaso), mas mesmo amostras muito grandes podem não ser representativas se o método de amostragem for defeituoso. Isso é chamado de **viés de amostragem**.




### Dados de Baixa Qualidade

- Obviamente, se seus dados de treinamento estiverem cheios de erros, discrepâncias e ruídos (por exemplo, devido a medições de baixa qualidade), será mais difícil para o sistema detectar os padrões subjacentes, portanto, é menos provável que seu sistema tenha um bom desempenho. Muitas vezes, vale a pena o esforço de gastar tempo limpando seus dados de treinamento. A verdade é que a maioria dos cientistas de dados passa uma parte significativa de seu tempo fazendo exatamente isso. Por exemplo:


- Se algumas **_instâncias_** forem claramente discrepantes, pode ser útil descartá-las ou tentar corrigir os erros manualmente.


- Se algumas **_instâncias_** estiverem faltando alguns **_features_** (por exemplo, 5% de seus clientes não especificaram sua idade), você deve decidir se deseja ignorar esse **_atributo_** completamente, ignorar essas **_instâncias_**, preencher os valores ausentes (por exemplo, com a mediana) ou treine um modelo com o **_feature_** e um modelo sem ele, e assim por diante.

### Features Irrelevantes 

- Como diz o ditado: lixo dentro, lixo fora. Seu sistema só será capaz de aprender se os **dados de treinamento** contiverem **_features_** relevantes suficientes e não muitos irrelevantes. Uma parte crítica do sucesso de um projeto de Machine Learning é apresentar um bom conjunto de **_features_** para você treinar. Esse processo, chamado de **_engenharia de feature_**, envolve:

  - **_Seleção de feature_**: selecione os **_features_** mais úteis para treinar entre os features existentes.
  - **_Extração de feature_**: combinando **_features_** existentes para produzir um mais útil (como vimos anteriormente, _algoritmos de redução de dimensionalidade_ podem ajudar).
  - Criando novos **_features_** reunindo novos dados.

### Overfitting nos Dados de Treinamento

- Digamos que você esteja visitando um país estrangeiro e o motorista do táxi o engane. Você pode ficar tentado a dizer que todos os taxistas daquele país são ladrões. A generalização excessiva é algo que nós, humanos, fazemos com muita frequência e, infelizmente, as máquinas podem cair na mesma armadilha se não tomarmos cuidado. No Machine Learning, isso é chamado de **overfitting** (**sobreajuste**): significa que o modelo tem um bom desempenho nos dados de treinamento, mas não generaliza bem.


- A (Figura 1-22) mostra um exemplo de um modelo de satisfação polinomial de alto grau que superajusta fortemente os dados de treinamento. Mesmo que ele tenha um desempenho muito melhor nos dados de treinamento do que no modelo linear simples, você realmente confiaria em suas previsões?

```python
Image('figura1-22.png')
```

 - Modelos complexos, como redes neurais profundas, podem detectar padrões sutis nos dados, mas se o conjunto de treinamento for ruído ou se for muito pequeno (o que introduz ruído de amostragem), é provável que o modelo detecte padrões no próprio ruído. Obviamente esses padrões não serão generalizados para novas instâncias:
 
   - Por exemplo, digamos que você alimente seu modelo de **satisfação da vida** com muitos outros **_atributos_**, incluindo os não informativos, como o nome do país. Nesse caso, um modelo complexo pode detectar padrões, como o fato de que todos os países nos dados de treinamento com **W** no nome têm uma satisfação com a vida maior que 7: New Zealand (7.3), Norway (7.4), Sweden (7.2), and Switzerland (7.5).
      
   - Qual é o seu grau de confiança de que a regra da satisfação-**W** generalize para Rwanda ou Zimbabwe? Obviamente, esse padrão ocorreu nos dados de treinamento por puro acaso, mas o modelo não tem como saber se um padrão é real ou simplesmente o resultado de ruído nos dados.
   

- O **overfitting** acontece quando o modelo é muito complexo em relação à quantidade e ruído dos dados de treinamento. As soluções possíveis são:
  
  - Simplificar o modelo selecionando um menor número de parâmetros (por exemplo, um modelo linear em vez de um modelo polinomial de alto grau), reduzindo o número de **_atributos_** nos dados de treinamento ou restringindo o modelo.
  - Reunir mais dados de treinamento.
    - Reduzir ruído nos dados de treinamento (corrigir erros nos dados e remover **_outliers_**).
    

- Restringir um modelo para simplificá-lo e reduzir o risco de **overfitting** é chamado **_regularização_**.

  - Por exemplo, o *modelo linear* que definimos anteriormente tem dois parâmetros, $\theta_0$ e $\theta_1$. Isso dá ao algoritmo de aprendizado dois graus de liberdade para adaptar o modelo aos dados de treinamento: ele pode ajustar a altura ($\theta_0$) e a inclinação ($\theta_1$) da linha. Se forçarmos $\theta_0 = 0$, o algoritmo terá apenas um grau de liberdade e terá muito mais dificuldade em ajustar os dados corretamente: tudo o que você pode fazer é mover a linha para cima ou para baixo (pois é o intercepto da equação) para se aproximar o máximo possível das instâncias de treinamento, então acabaria em torno da média.
  - Um modelo muito simples, de fato! Se permitirmos que o algoritmo modifique $\theta_1$, mas forçá-lo a mantê-lo pequeno, o algoritmo de aprendizado terá efetivamente algo entre um e dois graus de liberdade. Ele produzirá um modelo mais simples do que com dois graus de liberdade, porém mais complexo do que com apenas um grau de liberdade. Você deseja encontrar o equilíbrio certo entre ajustar perfeitamente os dados de treinamento e manter o modelo simples o suficiente para garantir que ele se generalize bem.
  

- A (Figura 1-23) mostra três modelos: a **linha pontilhada** representa o modelo original que foi treinado com alguns países ausentes, a **linha tracejada** é o nosso segundo modelo treinado com todos os países e a **linha sólida** é um modelo linear treinado com os mesmos dados que o primeiro modelo, mas com uma **_restrição de regularização_**. Você pode ver que a regularização forçou o modelo a ter uma inclinação menor, o que se encaixa um pouco menos nos dados de treinamento nos quais o modelo foi treinado, mas na verdade permite generalizar melhor para novos exemplos.

```python
Image("figura1-23.png")
```

- A quantidade de regularização a ser aplicada durante o aprendizado pode ser controlada por um **hiperparâmetro**. Um **hiperparâmetro** é um parâmetro de um algoritmo de aprendizado (**_não do modelo_**). Como tal, não é afetado pelo próprio algoritmo de aprendizado; deve ser **_definido antes do treinamento_** e permanece constante durante o treinamento. Se você definir o hiperparâmetro de regularização para um valor muito grande, obterá um modelo quase plano (uma inclinação próxima a zero); o algoritmo de aprendizado quase certamente não superestimará os dados de treinamento, mas será menos provável que encontre uma boa solução. Os hiperparâmetros de ajuste são uma parte importante da construção de um sistema de Machine Learning (você verá um exemplo detalhado no próximo capítulo).

 ### Underfitting nos Dados de Treinamento
 
- Como você pode imaginar, o **underfitting** é o oposto do **overfitting**: ocorre quando o modelo é muito simples para aprender a estrutura subjacente dos dados. Por exemplo, um modelo linear de satisfação da vida é propenso a ser **underfittting** (subajustado); a realidade é apenas mais complexa do que o modelo; portanto, suas previsões são imprecisas, mesmo nos exemplos de treinamento.

  - As principais soluções para corrigir esses problemas são:
    - Selecionando um modelo mais poderoso, com mais parâmetros.
    - Alimentando melhor os **features** ao algoritmo de aprendizado (engenharia de recursos).
    - Reduzir as restrições no modelo (por exemplo, reduzir o **_hiperparâmetro de regularização_**).

### RECAPTULANDO 


Até agora você já sabe muito sobre o **Machine Learning**. No entanto, passamos por tantos conceitos que você pode estar se sentindo um pouco perdido. Vamos dar um passo para trás e olhar para o cenário geral:

- O Machine Learning é sobre como melhorar as máquinas em alguma tarefa, aprendendo com os dados, em vez de precisar codificar regras explicitamente.


- Existem muitos tipos diferentes de sistemas de ML: **supervisionados** ou **não-supervionado**, **aprendizado em lote** ou **on-line**, **aprendizado baseados em instância** ou **modelo** e assim por diante.


- Em um projeto de ML, você **coleta dados** em um **conjunto de treinamento** e **alimenta o conjunto de treinamento com um algoritmo de aprendizado**. Se o algoritmo for **baseado em modelo**, ele ajusta alguns parâmetros para ajustá-lo ao conjunto de treinamento (ou seja, para fazer boas previsões no próprio conjunto de treinamento) e, esperançosamente, também será capaz de fazer boas previsões em novos casos. Se o algoritmo é **baseado em instância**, ele apenas aprende os exemplos memorizando e generalizando para novas instâncias comparando-as às instâncias aprendidas usando uma medida de similaridade.


- O sistema não funcionará bem se o seu conjunto de treinamento for muito pequeno ou se os dados não forem representativos, ruídos ou poluídos com **_features_** irrelevantes (entrada de lixo, saída de lixo). Por fim, seu modelo não precisa ser muito simples (nesse caso, será **underfitting**) nem muito complexo (nesse caso, será **0verfitting**).

Há apenas um **último tópico importante** a ser abordado: depois de treinar um modelo, você não quer apenas "esperar" que ele generalize para novos casos. Você deseja **avaliá-lo** e torná-lo **bem ajustado**, se necessário. Vamos ver como:

### Testando e Validando 

- A única maneira de saber até que ponto um modelo será generalizado para novos casos é realmente testá-lo em novos casos. Uma maneira de fazer isso é colocar seu modelo em produção e monitorar o desempenho dele. Isso funciona bem, mas se o seu modelo é horrivelmente ruim, os usuários irão reclamar --- não é a melhor ideia.


- Uma melhor opção é dividir seus dados em dois conjuntos: o **conjunto de treinamento** (**training set**) e o **conjunto de teste** (**test set**). Como esses nomes sugerem, você treina seu modelo usando o conjunto de treinamento e o testa usando o conjunto de teste. A **taxa de erro** em novos casos é chamada de **erro de generalização** (ou **erro fora da amostra**) e, ao avaliar seu modelo no conjunto de testes, você obtém uma estimativa desse erro. Esse valor informa o desempenho do modelo em **_instâncias_** (novas instâncias adicionadas ao banco de dados) nunca vistas antes.


- Se o erro de treinamento for baixo (ou seja, seu modelo cometer poucos erros no conjunto de treinamento), mas o erro de generalização for alto, isso significa que seu modelo está **sobreajustando** os dados de treinamento.

  - É comum usar **80% dos dados para treinamento** e reter **20% para testes**. No entanto, isso depende do tamanho do conjunto de dados: se ele contém 10 milhões de instâncias, reter 1% significa que seu conjunto de testes conterá 100.000 instâncias: provavelmente é mais do que suficiente para obter uma boa estimativa do erro de generalização.

### Ajuste de hiperparâmetro e seleção de modelo

- Portanto, avaliar um modelo é bastante simples: basta usar um conjunto de testes. Agora, suponha que você esteja hesitando entre dois modelos (digamos, um modelo linear e um modelo polinomial): como você pode decidir? Uma opção é treinar os dois e comparar o quanto eles generalizam usando o conjunto de testes.


- Agora, suponha que o modelo linear se generalize melhor, mas você deseja aplicar alguma regularização para evitar o ajuste excessivo. A questão é: como você escolhe o valor do hiperparâmetro de regularização? Uma opção é treinar 100 modelos diferentes usando 100 valores diferentes para esse hiperparâmetro. Suponha que você encontre o melhor valor de hiperparâmetro que produz um modelo com o menor erro de generalização, digamos apenas 5% de erro.


- Portanto, você lança esse modelo em produção, mas infelizmente ele não apresenta o desempenho esperado e produz 15% de erros. O que acabou de acontecer?


- O problema é que você mediu o erro de generalização várias vezes no conjunto de teste e adaptou o modelo e os hiperparâmetros para produzir o melhor modelo para esse conjunto específico. Isso significa que é improvável que o modelo tenha bom desempenho em novos dados.


- Uma solução comum para esse problema é chamada **holdout validation**: você simplesmente retém parte do conjunto de treinamento para avaliar vários modelos candidatos e selecionar o melhor. O novo conjunto de _holdout_ é chamado de **validation set** (ou, às vezes, **development set** ou de **dev set**). Mais especificamente, você treina vários modelos com vários hiperparâmetros no conjunto de treinamento reduzido (ou seja, o conjunto de treinamento completo menos o conjunto de validação) e seleciona o modelo com melhor desempenho no conjunto de validação. Após esse processo de **holdout validation**, você treina o melhor modelo em todo o conjunto de treinamento (incluindo o **validation set**), e isso fornece o modelo final. Por fim, você avalia esse modelo final em um **conjunto de testes** para obter uma estimativa do **erro de generalização**.


- Essa solução geralmente funciona muito bem. No entanto, se o **validation set** for muito **_pequeno_**, as avaliações do modelo serão imprecisas: você pode acabar selecionando um modelo abaixo do ideal por engano. Por outro lado, se o **validation set** for muito **_grande_**, o conjunto de treinamento restante será muito menor que o conjunto de treinamento completo. Por que isso é ruim? Bem, como o modelo final será treinado em todo o conjunto de treinamento, não é ideal comparar modelos candidatos treinados em um conjunto de treinamento muito menor. Seria como selecionar o velocista mais rápido para participar de uma maratona. 


- Uma maneira de resolver esse problema é executar a **cross-validation** validação cruzada repetida, usando muitos **validation sets** pequenos. Cada modelo é avaliado uma vez por **validation set**, depois é treinado no restante dos dados. Ao calcular a média de todas as avaliações de um modelo, obtemos uma medida de acurácia muito mais precisa de seu desempenho. No entanto, há uma desvantagem: o tempo de treinamento é multiplicado pelo número de **validation set**.

### Incompatibilidade de dados

- Em alguns casos, é fácil obter uma grande quantidade de dados para treinamento, mas não é perfeitamente representativo dos dados que serão usados na produção. Por exemplo, suponha que você queira criar um aplicativo móvel para tirar fotos de flores e determinar automaticamente suas espécies. Você pode facilmente baixar milhões de fotos de flores na Web, mas elas não serão perfeitamente representativas das fotos que serão realmente tiradas usando o aplicativo em um dispositivo móvel. Talvez você tenha apenas 10.000 fotos representativas (ou seja, realmente tiradas com o aplicativo). 


- Nesse caso, a regra mais importante a ser lembrada é que o **validation set** o **test set** devem ser o mais representativos possível dos dados que você espera usar na produção; portanto, eles devem ser compostos exclusivamente de imagens representativas: você pode embaralhá-las e colocar metade no **validation set** e metade no **test set** (certificando-se de que nenhuma duplicata ou quase duplicata termine em ambos os conjuntos).


- Depois de treinar seu modelo nas imagens da web, se você observar que o desempenho do seu modelo no **validation set** é decepcionante, não saberá se isso ocorre porque seu modelo teve **overfit** no **training set** ou se isso se deve apenas à incompatibilidade entre as imagens da Web e as imagens de aplicativos para dispositivos móveis.


- Uma solução é reter parte das imagens de treinamento (da Web) em outro conjunto que Andrew Ng chama de **train-dev set** (**conjunto de treinamento e desenvolvimento**). Depois que o modelo é treinado (no **training set**, **_NÃO_** no **conjunto de treinamento e desenvolvimento**), é possível avaliá-lo no **train-dev set**: se ele tiver um bom desempenho, o modelo não estará **sobreajustando** o **training set**, portanto, se tiver um desempenho ruim no **validation set**, o problema deve vir da incompatibilidade de dados.


- Você pode tentar resolver esse problema reprocessando as imagens da Web para torná-las mais parecidas com as fotos que serão tiradas pelo aplicativo móvel e, em seguida, treinando novamente o modelo. Por outro lado, se o modelo tiver um desempenho ruim no conjunto de **train-dev set**, o modelo poderá ter **sobreajuste** no **training set**, portanto, você deve tentar simplificar ou regularizar o modelo, obter mais dados de treinamento e limpar os dados de treinamento, conforme discutido anteriormente.

## No Free Lunch Theorem

- Um modelo é uma versão simplificada das observações. As simplificações destinam-se a descartar os detalhes supérfluos que dificilmente serão generalizados para novas instâncias. No entanto, para decidir quais dados descartar e quais manter, você deve fazer suposições. Por exemplo, um modelo linear assume que os dados são fundamentalmente lineares e que a distância entre as instâncias e a linha reta é apenas ruído, que pode ser ignorado com segurança.


- Em um famoso artigo de 1996, David Wolpert demonstrou que, se você não faz absolutamente nenhuma suposição sobre os dados, não há razão para preferir um modelo a outro. Isso é chamado de teorema **No Free Lunch** (**NFL**).

- Para alguns conjuntos de dados, o melhor modelo é um modelo linear, enquanto para outros conjuntos de dados é uma rede neural. Não há modelo que é a priori garantido para funcionar melhor (daí o nome do teorema). A única maneira de saber com certeza qual modelo é melhor é avaliar todos eles. Como isso não é possível, na prática, você fazer algumas suposições razoáveis sobre os dados e avaliar apenas alguns modelos razoáveis. Por exemplo, para tarefas simples, você pode avaliar modelos lineares com vários níveis de regularização e, para um problema complexo, você pode avaliar várias redes neurais.
