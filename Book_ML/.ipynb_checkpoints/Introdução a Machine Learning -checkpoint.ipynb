## Measuring Success: Training and Testing Data

- The objective is predict the species of iris for a new set of measurements through machine learning.
- But before we need  to know whether  it actually works -- that is, whether we should trust its predictions.


- Unfortunately, we can't use the completed model to evaluate it. Because it can remembr the whole training set, and therefore always predict the correct label for any point in the training set. 
- This "remembering" does not indicate *generalize* well of the model (or peformance well on new data). 


- To assess the model's performance, we show it new data (data that it hasn't seen before) fo which have labels.
- We will go split the data into two parts:
  - One part is used to build machine learning model -- called the  **_training data_** or **_training set_**.
  - Second part is used to assess how well the model works -- called **_test data_**, **_test set_**, or **_hold-out set_**.

- O **_scikit-learn_** tem uma função (**_train_test_split_**) que embaralha os dados e divide ele em duas partes:
  - A função extrai 75% das linhas para **_training set_**.
  - A função extrai 25% das linhas para **_test set_**.
- Pode ser escolhido diferentes percentuais para **_training set_** e **_test set_**, contudo, o percentual escolhido pela função é uma boa regra de ouro.
- Em **_scikit-learn_**, usualmente, os dados são denotados por $X$ maiúsculo, enquanto os rótulos são indicados por um $y$ minúsculo.


- Vamos chamar de **_train_test_split_** nossos dados e atribuir as saídas usando essa nomeclatura.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'],
                                                    random_state=0)
```

- Antes de fazer a divisão, a função **_train_test_split_** embaralha os dados usando um gerador de números pseudo-aleatórios.
  - Se apenas pegássemos os ultimos 25% dos dados para o conjunto de testes, teríamos os mesmos pontos de testes com rótulo 2 ( ver iris_dataset['target']) -- isso não permitiria generalizar o modelo. Dessa forma, é importante embaralhar os dados para que tenhamos todas as classes no conjunto de testes dos dados.

- Usamos o **_random_state_** para fixar uma semente fixa, pois, caso contrário, toda vez que gerarmos o código iremos encontrar pseudos números aleatórios diferentes. Daqui para frente, vamos sempre fixar esse número **zero** no **_random_state_**.

- O produto do **_train_test_split_** são todos Numpy arrays (*X_train*, *X_test*, *y_train*, *y_tets*) 
  - *X_train_* contém 75% das linhas dos dados, e *X_test* contém as outros 25% das linhas dos dados.

```python
print(f'X_train shape: {X_train.shape}')  ## 75% dos dados para treinamento -- fazer ML.  (caract. observáveis)

print(f'y_train shape: {y_train.shape}')  ## 75% dos targets para treinamento -- fazer ML  (var resultado)
```

```python
print(f'X_test shape: {X_test.shape}')  ## 25% dos dados para testes --
print(f'y_test shape: {y_test.shape}')  ## 25% dos targets para testes --
```
